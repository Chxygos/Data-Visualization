## Data Visualization

### Wrangling Process

In the analysis below we explore, clean where necessary our dataset so as to communicate using plot and graph our finding and for better understanding our analysis. 

The dataset set below is the includes information about individual rides made in a bike-sharing system

```python
# importing all packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# load and viewing in the dataset into a pandas dataframe
trip = pd.read_csv('201902-fordgobike-tripdata.csv')
trip.head()
```

```python
# Veiwing the data samples and variable with the the general info of our dataset
print(trip.shape)
trip.info()
```

```python
# General statistics of the dataset
trip.describe()
```

```python
# Converting start_time and end_time to datetime
trip[['start_time','end_time']] = trip[['start_time','end_time']].apply(pd.to_datetime)
```

```python
# Creating new column of Day and Month using the start_time
trip['day'] = trip['start_time'].dt.day_name()
trip['month'] = trip['start_time'].dt.month_name()
trip['hour'] = trip['start_time'].dt.hour
```

```python
# converting user_type and member_gender to category
trip['user_type'] = trip['user_type'].astype('category')
trip['member_gender'] = trip['member_gender'].astype('category')
```

```python
# checking for missing data
trip.isnull().sum()
```

```python
# Dropping the missing missing values in our dataset
trip = trip.dropna()
```

```python
# Calculating the individuals age and assigning it to a column
trip['age'] = 2019-trip['member_birth_year'].astype('int64')
```

```python
# dropping the member_birth_year
trip.drop('member_birth_year', axis=1, inplace=True)
```

```python
# Checking our dataseet info to confirm our changes 
trip.info()
```

```python
trip.head()
```

The dataset consists of missing values which were dropped, the age of individual users were calculate with the year in which the data were collected and the member_birth_year was dropped since it wasn't needed anymore during the wrangling process.
From the start_time we were able to create the day, hour and month column.

### Univariate Exploration

In the exploration of our data we will start by looking at the distribution of the Feature of Interest that is the duration_sec

```python
# Plotting a standard scale plot
bins = np.arange(0, trip['duration_sec'].max()+500, 500)

plt.figure(figsize=[8, 5])
plt.hist(data = trip, x = 'duration_sec', bins = bins)
plt.xlabel('Duration(sec)')
plt.show()
```

The distribution shows that there is a longtail, so we will try to put it together using a log scale

```python
# showing the statistics of duration_sec in log
np.log10(trip['duration_sec'].describe())
```

```python
# Showing the distribution in a log scale
bins = 10 ** np.arange(1.5, np.log10(trip['duration_sec'].max())+0.1, 0.1)

plt.figure(figsize=[8, 5])
plt.hist(data = trip, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([250, 500, 1e3, 2e3, 5e3, 1e4, 2e4], [250, 500, '1k', '2k', '5k', '10k', '20k'])
plt.xlabel('Duration(sec)')
plt.show()
```

Since the distribution has a long tail when plotting on a standard scale, we have to plot the duration_sec using a log-scale which looks as a unimodal, with more number of trips covered below 2000sec  and having it's peak around 600sec then starts to fall around 800sec to 20000sec

We will look at other features which can influence the duration_sec like distribution of the user_type,gender and age, then the start and end station_id

```python
# Plotting the user_type on a pie chart
plt.figure(figsize=[8,12])
sort_count = trip['user_type'].value_counts()
a_color = sb.color_palette()[0]
plt.subplot(2,1,1)
plt.pie(sort_count, labels= sort_count.index, startangle=90, counterclock=False, autopct='%.1f%%')
plt.title('A pie chart of the User Type')

# plotting gender on a bar chart
freq = trip['member_gender'].value_counts()
gen_order = freq.index
b_color = sb.color_palette()[1]
plt.subplot(2,1,2)
sb.countplot(data=trip, x='member_gender', color=b_color, order=gen_order)
plt.title('Distribution of the Genders');
```

From the pie chart of the user_type above we could see clearly that their is higher percentage of subscribers of about to 90.5% in comparison with 9.5% of customers users.
The second plot is a bar chart of member_gender, we observed as seen more number of males to compare with females and others in our dataset.

### Bivariate Exploration

Here we will view the relationship of one variable to another.

Let's start by viewing the coorelation between the duration_sec and age

```python
plt.figure(figsize=[8,5])
plt.scatter(data=trip, x='age', y='duration_sec', alpha=0.2 )
plt.axis([-5, 85, 500, 10000])
plt.xlabel('Age (years)')
plt.ylabel('Duaration(sec)')
plt.show()
```

The scatter plot of age on x-axis and duration on y-axis shows that more people between the age of 20 and 40 has higher duration of trip.

Let's view this in a heatmap for clearer understanding

```python
# heatmap of duration and age
bins_x = np.arange(0, trip['age'].max()+5, 5)
bins_y = np.arange(0, trip['duration_sec'].max()+500, 500)
plt.figure(figsize=[10,8])
plt.hist2d(data=trip, x='age', y='duration_sec', cmap='inferno', bins =[bins_x,bins_y])
plt.axis([0, 85, 100, 10000])
plt.colorbar()
plt.xlabel('Age(years)')
plt.ylabel('Duration(sec)')
plt.show()
```

Let's look at how the user_type affects the duration, then we can also investigate on the gender too 

```python
# Plotting a boxplot of user_type on duration
plt.figure(figsize=[8,4])
base_color = sb.color_palette()[0]
sb.boxplot(data=trip, x='user_type', y='duration_sec', color=base_color);
```

Let's rescale so we can see our values

```python
# Defining a function for plotting a boxplot
def boxgrid(x, y, **kwargs):
    base_color = sb.color_palette()[0]
    sb.boxplot(x,y, color=base_color)
    
# plotting a boxplot of user_type and member_gender on duration   
plt.figure(figsize= (5,5))
g = sb.PairGrid(data=trip, y_vars='duration_sec', x_vars=['user_type', 'member_gender'], height=6)
g.map(boxgrid)
plt.ylim([-5,2500])
plt.show();
```

In the first box plot of the user_type we observed that higher percentage of customers make longer trip in comparison with the subscribers.
The second boxplot is the plot of member_gender against the duration_sec, we can confirm that despite the percentage of male being the most users as seen in the distribution of the member_gender in the univariate exploration, when compare with the duration we see that we have more percentage of female and other riding longer than the male.

```python
# Plotting a cluster barchart of user_type and bike_share_for_all_trip
plt.figure(figsize=(8,5))
sb.countplot(data=trip, x='user_type', hue='bike_share_for_all_trip', palette='Blues');
```

We can see from the plot that bike share is only applicable to subscribers.

```python
# Plotting a cluster bar chart of the day and the usertype
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(10,8))
sb.countplot(data=trip, x='day', hue='user_type', order=days);
```

From the cluster bar chart above the graph behaves normal, the customer and subscribers have both peak on Thursdays.

```python
# # Plotting a cluster bar chart of the hour and the usertype
hours = np.arange(0,24,1)
plt.figure(figsize=(10,8))
sb.countplot(data=trip, x='hour', hue='user_type', order=hours);
```

Here in the cluster barchart of hour and usertype the customer appears to ride more during working hours.

### Multivariate Exploration

Here we will look the relation between more than two variables in our dataset.

```python
g = sb.FacetGrid(data = trip, col = 'user_type', col_wrap = 2, height = 5,
                 xlim = [-5, 85], ylim = [-500, 8000])
g.map(plt.scatter, 'age', 'duration_sec', alpha=0.25)
g.set_xlabels('Age')
g.set_ylabels('Duration')

plt.show()
```

In the both scatter plot, they are similar with the scatter plot in the bivariate section of age against duration which showed that longer duration are taken by younger people ranging from 20 to 40years. The Customers user_type says the same but the Subscribers has showed an extention in age to 60years 

```python
# Plotting a cluster barplot for hour, duration and user type
plt.figure(figsize=(10,8))
sb.barplot(data=trip, x='hour', y='duration_sec', hue='user_type',ci=None, order=hours);
```

From the plot, the customers have closely a consistent trip duration while the subscribers flutuates. 

```python
# Plotting a pointplot for day, duration and user type
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(8,5))
sb.pointplot(data=trip, x='day', y='duration_sec', hue='user_type',ci=None, order=days);
```

From the pointplot, we see the customers spend more duration with most duration during weekends. The subscribers seems to be consistent in their duration with little increase on weekends.

```python
# Plotting a pointplot for member_gender, duration and user type
plt.figure(figsize=(8,5))
ax = sb.pointplot(data=trip, x='member_gender', y='duration_sec', hue='user_type', palette='Blues', linestyles='', dodge = 0.4)
ax.set_yticklabels([],minor = True);
```

The other gender seems to spend more duration on trips for both user types, then the females and the males that appears to spend least duration.
